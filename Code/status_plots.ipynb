```python
def drawchart(coords):
    import matplotlib.pyplot as plt
    from matplotlib.path import Path
    import matplotlib.patches as patches

    verts = [
        (-4., 4.), # left, bottom
        (4., 4.), # left, top
        (4., -4.), # right, top
        (-4., -4.), # right, bottom
        (0., 0.), # ignored
        ]

    codes = [Path.MOVETO,
             Path.LINETO,
             Path.LINETO,
             Path.LINETO,
             Path.CLOSEPOLY,
             ]

    path = Path(verts, codes)

    fig = plt.figure()
    ax = fig.add_subplot(111)
    N = 5
    x = coords[0]
    y = coords[1]
    #patch = patches.PathPatch(path, facecolor='#cceeff', lw=2)
    #ax.add_patch(patch)
    rect = patches.Rectangle((-4,-4),8,8,linewidth=1,edgecolor='b',facecolor='#cceeff', zorder=0)
    rect2 = patches.Rectangle((-0.5,-0.5),1,1,linewidth=1,edgecolor='b',facecolor='#0000ff', zorder=0)

    # Add the patch to the Axes
    ax.add_patch(rect)
    ax.add_patch(rect2)


    ax.scatter(x, y, marker='x', color='r')

    ax.set_xlim(-10,10)
    ax.set_ylim(-10,10)
    plt.show()
```

```python
def unzip(roundlist):
    x = [i[0] for i in roundlist]
    y = [i[1] for i in roundlist]
    return [x,y]
```
