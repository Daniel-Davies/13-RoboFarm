```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_data = [1/8,
              1/8,
              0/8,
              1/10,
              0/10,
              2/10,
              0/6,
              1/6,
              1/10,
              1/8]
```

```python
genetic_data = [4/8,
               3/8,
               4/8,
               5/10,
               3/10,
               6/10,
               4/6,
               3/6,
               4/8,
               6/10]
```

```python
trial_nums = [num for num in range(1,11)]
```

```python
plt.figure(figsize=(8,6))

plt.plot(trial_nums, random_data, '-o', c='red')
plt.plot(trial_nums, genetic_data, '-o', c='blue')
plt.xlabel("Trial Number", fontsize=20)
plt.ylabel("Yield Percentage", fontsize=20)
plt.title("Yield Graph", fontsize=30)

plt.legend(['Baseline', 'Genetic'])
plt.show()
```
